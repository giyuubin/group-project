```python
import time
import pandas as pd
from sklearn.metrics import accuracy_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
# K-NN은 실시간 탐지에 부적합하므로 통상적으로 EDR 벤치마크에서 제외하거나 비교용으로만 둔다

# 1. 평가할 모델들 딕셔너리로 준비 (기본 하이퍼파라미터)
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM (RBF)": SVC(kernel='rbf', random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "DNN (MLP)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# 결과를 저장할 리스트
results = []

print("🛡️ 악성코드 탐지 모델 교차 검증(Cross-validation)을 시작합니다...\n")

# 2. 각 모델별로 학습 및 평가 진행
for model_name, model in models.items():
    print(f"[{model_name}] 모델 학습 및 추론 중...")

    # 훈련 (이미 SMOTE 등이 적용된 X_train, y_train을 사용한다고 가정)
    # model.fit(X_train, y_train)  <-- 실제 데이터로 주석 해제

    # 추론 속도 측정 시작
    start_time = time.time()

    # 예측 (1:23 불균형인 원본 X_test 사용)
    # y_pred = model.predict(X_test) <-- 실제 데이터로 주석 해제

    # 추론 속도 측정 종료 (밀리초 단위 변환)
    end_time = time.time()
    # inference_time_ms = (end_time - start_time) * 1000 / len(X_test)

    # 임시 가짜 데이터 (실제 실행 시 위 코드의 주석을 풀고 아래 3줄은 지우세요)
    acc = 0.96 # accuracy_score(y_test, y_pred)
    recall = 0.85 # recall_score(y_test, y_pred)
    inference_time_ms = 5.0

    # 결과 저장
    results.append({
        "Model": model_name,
        "Accuracy (%)": round(acc * 100, 2),
        "Recall (%)": round(recall * 100, 2),
        "Inference Time (ms/sample)": round(inference_time_ms, 3)
    })

# 3. 결과를 데이터프레임으로 변환하여 표 형태로 출력
benchmark_df = pd.DataFrame(results)
benchmark_df.set_index("Model", inplace=True)

print("\n📊 [모델별 탐지 성능 및 추론 속도 벤치마크 결과]")
print(benchmark_df.to_markdown())

# 향후 이 benchmark_df의 수치들을 마크다운 문서의 표와 그래프 값에 그대로 덮어씌우면 됨
```
