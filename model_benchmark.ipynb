```python
import time
import pandas as pd
# 정밀도와 F1-Score 메트릭 추가
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

# 0. 실제 데이터 로드 (Phase 3 파일 기준)
# 실제 파일 경로에 맞게 수정해주세요 (예: 'smote_train_data.csv', 'optimized_data_test.csv')
print("데이터를 불러오는 중입니다...")
# X_train = pd.read_csv('smote_train_data.csv').drop(columns=['malware'])
# y_train = pd.read_csv('smote_train_data.csv')['malware']
# X_test = pd.read_csv('test_data.csv').drop(columns=['malware'])
# y_test = pd.read_csv('test_data.csv')['malware']

# 1. 평가할 모델들 딕셔너리로 준비 (랜덤 포레스트 OOB 추가)
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM (RBF)": SVC(kernel='rbf', random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "DNN (MLP)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, oob_score=True, random_state=42) # oob_score 추가됨
}

results = []
print("악성코드 탐지 모델 교차 검증(Cross-validation)을 시작합니다...\n")

# 2. 각 모델별로 학습 및 평가 진행
for model_name, model in models.items():
    print(f"[{model_name}] 모델 학습 및 추론 중...")

    # 훈련 (실제 데이터 준비 후 아래 주석 해제)
    # model.fit(X_train, y_train)

    start_time = time.time()
    # 예측 (실제 데이터 준비 후 아래 주석 해제)
    # y_pred = model.predict(X_test)
    end_time = time.time()

    # 추론 속도 계산 (1개 샘플당 밀리초)
    # inference_time_ms = (end_time - start_time) * 1000 / len(X_test)

    # --- 아래 임시 데이터는 실제 실행 시 삭제하고 실제 메트릭으로 교체하세요 ---
    # acc = accuracy_score(y_test, y_pred)
    # recall = recall_score(y_test, y_pred)
    # precision = precision_score(y_test, y_pred)
    # f1 = f1_score(y_test, y_pred)
    acc, recall, precision, f1, inference_time_ms = 0.96, 0.85, 0.90, 0.87, 5.0
    # -------------------------------------------------------------------------

    results.append({
        "Model": model_name,
        "Accuracy (%)": round(acc * 100, 2),
        "Recall (%)": round(recall * 100, 2),
        "Precision (%)": round(precision * 100, 2), # 추가됨
        "F1-Score": round(f1, 4),                   # 추가됨
        "Inference Time (ms/sample)": round(inference_time_ms, 3)
    })

# 3. 결과 출력
benchmark_df = pd.DataFrame(results)
benchmark_df.set_index("Model", inplace=True)
print("\n📊 [모델별 탐지 성능 및 추론 속도 벤치마크 결과]")
print(benchmark_df.to_markdown())
```
